```python
keys = ['a', 'b', 'c']
values = [1, 2, 3]

def merge_lists_to_dictionary(keys, values):
    return {keys[i]: values[i] for i in range(min(len(keys), len(values)))}

merge_lists_to_dictionary(keys,values)
```
