```python
'''
Parameters:

list1 (List of integers): The first sorted list.

list2 (List of integers): The second sorted list.

Returns:

A single list of integers, containing all elements from list1 and list2, sorted in non-decreasing order.

'''
```

```python
list1 = [1,3,5]
list2 = [2,4,6]

# first thought

def merge_two_sorted_lists(list1, list2):
    
    result = list1+list2
    result.sort()
    return result

merge_two_sorted_lists(list1,list2)
```

```python
# further try

list1 = [1,3,5]
list2 = [2,4,6]

# brute force
def merge_two_sorted_listss(list1, list2):
    result = list1 + list2
    
    for i in range(len(result)):
        for j in range(i + 1, len(result)):
            if result[i] > result[j]:
                result[i], result[j] = result[j], result[i]
    
    return result

merge_two_sorted_listss(list1,list2)

# inefficient way
```

```python
list1 = [1,3,5]
list2 = [2,4,6]

def merge_2_sorted_list(list1,list2):
    result = []
    i = 0
    j = 0
    
    while i<len(list1) and j<len(list2):
        if list1[i]<=list2[j]:
            result.append(list1[i])
            i+=1
        else:
            result.append(list2[j])
            j+=1
            
    # adding remaining elements to list
    
    while i<len(list1):
        result.append(list1[i])
        i+=1
        
    while j<len(list2):
        result.append(list2[j])
        j+=1
        
    return result

merge_2_sorted_list(list1,list2)

# optimal solution
```
