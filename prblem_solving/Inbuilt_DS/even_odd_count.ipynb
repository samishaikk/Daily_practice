```python
def count_even_odd(lst):
    # Your code goes here
    # output = []
    
    even_count = 0
    odd_count = 0
    
    for num in lst:
        if num%2==0:
            even_count+=1
        else:
            odd_count+=1
            
    return even_count,odd_count

count_even_odd(lst=[1,2,3,4,5,6,7])
```
