```python
# pop and insert method
def rotate_lst(lst,k):
    
    if len(lst)==0:
        return lst
    
    k = k % len(lst)
    
    for _ in range(k):
        last = lst.pop()
        lst.insert(0,last)
        
    return lst

rotate_lst(lst=[1,2,3,4,5,56,67],k=3)
```

```python
# three reversal

def rotate_list(lst,k):
    
    if len(lst)==0:
        return lst
    
    k = k % len(lst)
    n = len(lst)
    
    def reverse(start,end):
        while start<end:
            lst[start],lst[end] = lst[end],lst[start]
            start += 1
            end -= 1
        
    # first, have to reversing entire list
    reverse(0,n-1)
    
    # then, reverse first k elements
    reverse(0,k-1)
    
    # reversing remaining elements
    reverse(k,n-1)
    
    return lst

rotate_list(lst=[1,2,3,4,5,6,7],k=3)
    
```
