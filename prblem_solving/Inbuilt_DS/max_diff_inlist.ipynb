```python
# Maximum Difference between 2 consecutive elements in a List
# Find Maximum Difference Between Two Consecutive Elements (Brute Force Approach)
```

```python
def max_consecutive_difference(lst):
    
    if len(lst)<2:
        return 0
    
    max_diff = 0
    
    for i in range(len(lst) -1):
        diff = abs(lst[i]-lst[i+1])
        if diff>max_diff:
            max_diff = diff
            
    return max_diff
        
        
    # largest_elem = max(lst)
    # smallest_elem = min(lst)
    
    # max_diff = largest_elem-smallest_elem
    
    # return max_diff

print(f'max diff between the list is: {max_consecutive_difference(lst=[1.48,23,44,93,8,28])}')
```
