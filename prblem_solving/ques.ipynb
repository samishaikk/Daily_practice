```python
# Write a program to calculate the factorial of a number using a loop

num = int(input('input: '))


def factorial(num):
    result = 1
    if num<0:
        return "no factorial for negative"
    elif num ==0 or num ==1:
        return 1
    elif num>2:
        for i in range(2,num+1):
            result *= i
        return result
    
factorial(num)
```

```python
# Check if a number is a palindrome

given_num = str(int(input('enter num: ')))

given_num = ''.join(c.lower() for c in given_num if c.isalnum())
if given_num[::] == given_num[::-1]:
    print('palindrome number')
else:
    print('not a palindrome')
```

```python
# Check if List is Subset of Another List

def is_subset(lst1, lst2):
    # Your code goes here
    lst1 = [1, 2, 3]
    lst2 = [1, 2, 3, 4, 5]
    
    for i in range(len(lst1)):
        found = False
        
        for j in range(len(lst2)):
            if lst1[i]==lst2[i]:
                found = True
                break
        
        if not found:
            return False
        
        
    return True
```

```python
# Create a program to print all Armstrong numbers between 1 to 1000

def is_armstrong(number):
   
   num_str = str(number)
   
   num_digits = len(num_str)
   
   sum_of_powers = sum(int(digit) ** num_digits for digit in num_str)
   
   return sum_of_powers == number

print(is_armstrong(153)) 
print(is_armstrong(1253)) 
```

```python
# Create a calculator app using if-else.

num1 = int(input('enter num: '))
num2 = int(input('enter num: '))
operator = input('enter operator(+,-,/,*): ')
if operator == '+':
    print(f'addition of {num1} and {num2} is {num1+num2}')
    
elif operator == '-':
    print('result: ',num1-num2)
    
elif operator == '*':
    print('result: ',num1*num2)
    
elif operator == '/':
    if num2!=0:
        print('result: ',num1/num2)
    else:
        print('division by zero is infinity')
        
else:
    print('invalid operator')
```

```python
# Convert a decimal number to binary using loops

n = int(input('user input: '))


if n == 0:
    print('0')
    
else:
    binary = ""
    while n > 0:
        remainder = n%2
        binary = str(remainder)+binary
        n = n//2

    print('binary: ',binary)
```

```python
# Create a program to check whether a number is prime or not

n = int(input('user input: '))

if n<=1:
    print('not prime')
    
else:
    is_prime = True
    
    for i in range(2,n):
        if n % i == 0:
            is_prime = False
            break
        
    if is_prime:
        print('prime number')
    else:
        print('not a prime')
```
