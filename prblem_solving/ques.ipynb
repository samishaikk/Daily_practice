```python
# Write a program to calculate the factorial of a number using a loop

num = int(input('input: '))


def factorial(num):
    result = 1
    if num<0:
        return "no factorial for negative"
    elif num ==0 or num ==1:
        return 1
    elif num>2:
        for i in range(2,num+1):
            result *= i
        return result
    
factorial(num)
```

```python
# Check if List is Subset of Another List

def is_subset(lst1, lst2):
    # Your code goes here
    lst1 = [1, 2, 3]
    lst2 = [1, 2, 3, 4, 5]
    
    for i in range(len(lst1)):
        found = False
        
        for j in range(len(lst2)):
            if lst1[i]==lst2[i]:
                found = True
                break
        
        if not found:
            return False
        
        
    return True
```

```python
# Create a program to print all Armstrong numbers between 1 to 1000

def is_armstrong(number):
   
   num_str = str(number)
   
   num_digits = len(num_str)
   
   sum_of_powers = sum(int(digit) ** num_digits for digit in num_str)
   
   return sum_of_powers == number

print(is_armstrong(153)) 
print(is_armstrong(1253)) 
```

```python
# Write a program to find the second highest number in a list

def secondhighest():
    
    givenlist = [2,3,4,5,6,7]

    first = second = float('-inf')

    for num in givenlist:

        if num > first:
            second = first
            first = num

        elif num > second and num != first:
            second = num

    print("Second Highest =", second)

secondhighest()
    
```

```python
# Create a function to rotate a list left by k positions

def rotate_left(arr, k):

    k = k % len(arr)

    return arr[k:] + arr[:k]


numbers = [1, 2, 3, 4, 5]

rotate_left(numbers, 2)
```

```python
# Write a function to find the missing number from a list of 1 to N

def find_missing(arr, n):

    expected_sum = n * (n + 1) // 2
    actual_sum = sum(arr)

    return expected_sum - actual_sum


numbers = [1, 2, 3, 5]

print(find_missing(numbers, 5))
```

```python
# Write a function to merge two dictionaries and handle key collisions by summing values

def merge_dicts(d1, d2):

    result = d1.copy()

    for key, value in d2.items():

        if key in result:
            result[key] += value

        else:
            result[key] = value

    return result


d1 = {'a': 10, 'b': 20, 'c': 30}
d2 = {'b': 5, 'c': 15, 'd': 40}

print(merge_dicts(d1, d2))
```

```python
# Create a function to find unique elements present in only one of two lists

def unique_elements(list1, list2):

    result = []

    for x in list1:
        if x not in list2:
            result.append(x)

    for x in list2:
        if x not in list1:
            result.append(x)

    return result

a = [1, 2, 3, 4]
b = [3, 4, 5, 6]

print(unique_elements(a, b))
```
