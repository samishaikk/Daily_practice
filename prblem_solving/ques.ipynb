```python
# Write a program to calculate the factorial of a number using a loop

num = int(input('input: '))


def factorial(num):
    result = 1
    if num<0:
        return "no factorial for negative"
    elif num ==0 or num ==1:
        return 1
    elif num>2:
        for i in range(2,num+1):
            result *= i
        return result
    
factorial(num)
```

```python
# Check if a number is a palindrome

given_num = str(int(input('enter num: ')))

given_num = ''.join(c.lower() for c in given_num if c.isalnum())
if given_num[::] == given_num[::-1]:
    print('palindrome number')
else:
    print('not a palindrome')
```

```python
# Program to calculate sum of digits using sum() and string conversion

num = int(input("Enter a number: "))

# Take absolute value to handle negative numbers
total = sum(int(digit) for digit in str(abs(num)))

print(f"Sum of digits of {num} is {total}")

```
