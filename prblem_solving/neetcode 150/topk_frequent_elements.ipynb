### heap 

 
1. COUNT frequencies
2. FIND top k by frequency -> heap (maintain top k)
3. RETURN the k numbers

```python
### heap solution

from collections import Counter

def topKFrequent(nums, k):
    
    count = Counter(nums)
    
    
    sorted_items = sorted(count.items(), key=lambda x: x[1], reverse=True)
    

    return [num for num, freq in sorted_items[:k]]

topKFrequent(nums=[1,2,2,3,3,3],k=2)
```

```python
### sorting hashmap SOLUTION
```

```python
def topKFrequent(nums, k):
    
    count = {}
    
    for num in nums:
        if num in count:
            count[num] += 1
        else:
            count[num] = 1
            
    sorted_items = sorted(count.items(), key=lambda x: x[1], reverse=True)
    
    return [num for num, freq in sorted_items[:k]]

topKFrequent(nums=[1,2,2,3,3,3],k=2)
```
