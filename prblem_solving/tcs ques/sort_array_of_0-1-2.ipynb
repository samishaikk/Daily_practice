#### Sort a list containing only 0s, 1s, and 2s in a single pass without using sort().
- Input:
[2,0,1,1,0,2]
- Output:
[0,0,1,1,2,2]

```python
### APPROACH 
```

- Use three pointers: lo=0, mid=0, hi=n-1.
- If a[mid]=0: swap(lo,mid), lo++, mid++.
- If a[mid]=1: mid++. If a[mid]=2: swap(mid,hi), hi--.
- Repeat while mid <= hi

```python
def sort012(arr):
    low = 0
    mid = 0
    high = len(arr) - 1
    
    while mid<= high:
        if arr[mid] == 0:
            arr[low], arr[mid] = arr[mid], arr[low]
            low+=1
            mid+=1
            
        elif arr[mid] == 1:
            mid+=1
            
        else:
            arr[mid],arr[high] = arr[high],arr[mid]
            high -=1
    return arr

print(sort012(arr=[2,0,1,1,0,2]))
```
