```python
def secondlargest(arr):
    if len(arr)<2:
        return None
    
    unique_arr = list(set(arr))
    if len(unique_arr)<2:
        return None
    unique_arr.sort()
    return unique_arr[-2]

secondlargest(arr=[5,2,8,1,9])
```
