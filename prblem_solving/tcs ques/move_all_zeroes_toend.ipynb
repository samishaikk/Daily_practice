
#### Given a list, move all 0s to the end while maintaining the relative order of non-zero elements.
- Input:
[0,1,0,3,12]
- Output:
[1,3,12,0,0]

```python
### APPROACH
```


- Use two-pointer: i=0 (slow), j iterates the list.
- When a[j] != 0, swap a[i] and a[j], then i++.
- After loop, all zeros are pushed to the end

```python
def move_zeroes(a):
    i = 0
    for j in range(len(a)):
        if a[j]!=0:
            a[i],a[j] = a[j],a[i]
            i+=1
    return a

print(move_zeroes([0,1,0,3,12]))
```
