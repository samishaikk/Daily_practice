### Types of statistics

- Descriptive & Inferential

- Descriptive -> Involves methods for Summarizing and organizing data to make it understandable. This types help us to describe the basic features of data in a study.

1. Measure of central tendency [mean,median,mode]
2. Measure of dispersions [variance,standard dev]
3. Data distribution -> histograms, boxplot, piechart, pdf & pmf
4. Summary statistics -> five number summary , Q1 Q2 Q3 maximum


```python
### Inferential 
```

- Inferential -> involves methods for making predictions or inferences about a population based on a sample of data. Allows for hypothesis testing, estimation, and drawing conclusions.

1. Hypothesis testing
2. P value
3. Confidence interval 
4. statistical analysis test -> z-test, t-test, ANOVA(f-test), chi-square
